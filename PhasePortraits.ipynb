```python
from phaseportrait import PhasePortrait2D
import numpy as np
```

```python
#Parameters from the equation, accessed by name
N = 100

p    = {
"alpha"  : 0.75,            # Per capita rate of spontaneous discoveries
"s"      : 3.5,             # Per capita rate of ant leaving trail per distance
"gamma1" : 0.2,             # Range of foraging scouts
"gamma2" : 0.021,           # Range of recruitment activity
"gamma3" : 0.021,           # Range of influence of pheromone
"K"      : 1,               # Inertial effects that may affect pheromones
"n1"     : 20,              # Individual ant's contribution to rate of recruitment (orig. eta1)
"n2"     : 20               # Pheromone strength of trail (originally eta2)       
}

Q = np.array([3, 15])
D = np.array([3, 10])
```

```python
def dF(x_a, x_b):
    betaB = p["n1"] * Q
    betaS = p["n2"] * Q
    da = (p["alpha"]* np.exp(-p["gamma1"]*D[0]) + (p["gamma2"]/D[0])*betaB[0]*x_a)*(N-x_a-x_b) - (p["s"]*D[0]*x_a)/(p["K"]+ (p["gamma3"]/D[0])*betaS[0]*x_a)
    db = (p["alpha"]* np.exp(-p["gamma1"]*D[1]) + (p["gamma2"]/D[1])*betaB[1]*x_b)*(N-x_a-x_b) - (p["s"]*D[1]*x_b)/(p["K"]+ (p["gamma3"]/D[1])*betaS[1]*x_b)
    return da, db

def dF_lim(x_a, x_b):
    da = x_a * Q[0]/D[0]
    db = x_b * Q[1]/D[1]
    return da, db
```

```python
%matplotlib widget
N=100
phaseportrait = PhasePortrait2D(dF, [0, N], Title = "Phase portrait of number of ants on each trail", xlabel="Trail 1", ylabel="Trail 2")
#phaseportrait.add_nullclines(xcolor='black', ycolor='green')
phaseportrait.plot()

N = 100000
phaseportrait = PhasePortrait2D(dF, [0, N], Title = "Phase portrait of number of ants on each trail", xlabel="Trail 1", ylabel="Trail 2")
#phaseportrait.add_nullclines(xcolor='black', ycolor='green')
phaseportrait.plot()

phaseportrait = PhasePortrait2D(dF_lim, [0, 1], Title = "Phase portrait of number of ants on each trail", xlabel="Trail 1", ylabel="Trail 2")
#phaseportrait.add_nullclines(xcolor='black', ycolor='green')
phaseportrait.plot()
```
